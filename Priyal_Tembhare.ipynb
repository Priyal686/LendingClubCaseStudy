```python
#Counting how many rows are fully paid,charged off and current in loan status column
count_loan_status=data['loan_status'].value_counts()
print(count_loan_status)
```

```python
#As a part of univarinet alalysis see how many are charged off,fully paid and current using graph
status_count=data['loan_status'].value_counts()

status_count.plot(kind='bar',color='skyblue',edgecolor='black')

plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.title('Histogram of Status Categories')
plt.show()
```

```python
#Doing bivarinet analysis to see comparision impact of loan_status in home_ownership
cross_table=pd.crosstab(data['home_ownership'],data['loan_status'])

cross_table.plot(kind='bar', stacked=True, colormap='viridis')

plt.xlabel('Home Owner')
plt.ylabel('Count')
plt.title('Comparison of Home Owner and Loan Status')
plt.legend(title='Home Owner vs Loan Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Checking relation between  Verification status and loan status
cross_table_verification=pd.crosstab(data['verification_status'],data['loan_status'])
cross_table_verification.plot(kind='bar', stacked=True, colormap='viridis')
plt.xlabel('Verification Status')
plt.ylabel('Count')
plt.title('Comparison of Verification Status and Loan Status')
plt.legend(title='Verification Status vs Loan Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#adding a new column to show onlu getting default and not get default
data['loan_status_0_1'] = data['loan_status'].map({'Fully Paid': 1, 'Current': 1, 'Charged Off': 0})
data
#Plotting box plot for loan status and annual income
plt.figure(figsize=(8, 6))
sns.boxplot(x='loan_status_0_1', y='dti', data=data, palette='viridis')

plt.xlabel('Loan Status')
plt.ylabel('dti')
plt.title('Box Plot of dti by loan status')
plt.show()
```

```python
#plotting box plot for loan status and open account
plt.figure(figsize=(8, 6))
sns.boxplot(x='loan_status_0_1', y='open_acc', data=data, palette='viridis')

plt.xlabel('Loan Status')
plt.ylabel('open account')
plt.title('Box Plot of Account open by loan status')
plt.show()
```

```python
#plotting pie chart 
fully_paid_df = data[data['home_ownership'] == 'OWN']

# Count the occurrences for each category of 'Home Owner'
home_owner_counts = fully_paid_df['loan_status'].value_counts()

# Create a pie chart
labels = home_owner_counts.index
sizes = home_owner_counts.values
colors = ['lightgreen', 'lightblue', 'lightcoral']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Fully Paid Loans by Home Owner ')
plt.show()

fully_paid_df = data[data['home_ownership'] == 'RENT']

# Count the occurrences for each category of 'Home Owner'
home_owner_counts = fully_paid_df['loan_status'].value_counts()

# Create a pie chart
labels = home_owner_counts.index
sizes = home_owner_counts.values
colors = ['lightgreen', 'lightblue', 'lightcoral']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Fully Paid Loans by Rent Category')
plt.show()

ully_paid_df = data[data['home_ownership'] == 'MORTGAGE']

# Count the occurrences for each category of 'Home Owner'
home_owner_counts = fully_paid_df['loan_status'].value_counts()

# Create a pie chart
labels = home_owner_counts.index
sizes = home_owner_counts.values
colors = ['lightgreen', 'lightblue', 'lightcoral']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Fully Paid Loans by Mortgage')
plt.show()
```
